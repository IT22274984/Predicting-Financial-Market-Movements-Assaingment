## 2. Deep Learning Model Design

### LSTM

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import mean_squared_error, r2_score

# === RESHAPE INPUTS FOR LSTM ===
X_train_lstm = np.expand_dims(X_train, axis=1)
X_test_lstm = np.expand_dims(X_test, axis=1)

# === CONVERT TO TENSORS ===
X_train_tensor = torch.FloatTensor(X_train_lstm)
X_test_tensor = torch.FloatTensor(X_test_lstm)
y_train_tensor = torch.FloatTensor(y_train.to_numpy()).reshape(-1, 1)
y_test_tensor = torch.FloatTensor(y_test.to_numpy()).reshape(-1, 1)

# === DEFINE DATA LOADERS ===
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size1=64, hidden_size2=32, dense_size=16):
        super(LSTMModel, self).__init__()
        self.lstm1 = nn.LSTM(input_size, hidden_size1, batch_first=True)  # Removed return_sequences
        self.dropout1 = nn.Dropout(0.2)
        self.lstm2 = nn.LSTM(hidden_size1, hidden_size2, batch_first=True)  # Removed return_sequences
        self.dropout2 = nn.Dropout(0.2)
        self.dense1 = nn.Linear(hidden_size2, dense_size)
        self.relu = nn.ReLU()
        self.dense2 = nn.Linear(dense_size, 1)

    def forward(self, x):
        x, _ = self.lstm1(x)
        x = self.dropout1(x)
        x, _ = self.lstm2(x)
        x = self.dropout2(x)
        x = x[:, -1, :]  # Get last time step
        x = self.dense1(x)
        x = self.relu(x)
        x = self.dense2(x)
        return x
```

```python
# Initialize model
input_size = X_train_lstm.shape[2]
model = LSTMModel(input_size)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters())

# Create DataLoader for batch processing
from torch.utils.data import TensorDataset, DataLoader

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
#Early stopping parameters
patience = 10
best_val_loss = float('inf')
epochs_no_improve = 0
best_model_state = None

# Training loop
num_epochs = 100
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    # Training
    model.train()
    train_loss = 0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    
    train_loss /= len(train_loader)
    train_losses.append(train_loss)
    
    #Validation
    model.eval()
    with torch.no_grad():
        # Use 20% of training data for validation
        val_size = int(0.2 * len(X_train_tensor))
        X_val = X_train_tensor[:val_size]
        y_val = y_train_tensor[:val_size]
        
        val_outputs = model(X_val)
        val_loss = criterion(val_outputs, y_val).item()
        val_losses.append(val_loss)
    
    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_model_state = model.state_dict()
        epochs_no_improve = 0
    else:
        epochs_no_improve += 1
    
    if epochs_no_improve >= patience:
        print(f'Early stopping triggered after epoch {epoch+1}')
        break
    
    print(f'Epoch {epoch+1}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')
```

```python
# Load best model
model.load_state_dict(best_model_state)
```

```python
# Evaluate model
model.eval()
with torch.no_grad():
    y_pred_tensor = model(X_test_tensor)
    y_pred = y_pred_tensor.numpy().flatten()

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"LSTM Model - MSE: {mse}")
print(f"LSTM Model - R2: {r2}")
```

```python
# Plot training history
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('LSTM Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### GRU Model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# === RESHAPE & CONVERT DATA ===
# Reshape for GRU: (batch_size, seq_len, features)
X_train_gru = np.expand_dims(X_train, axis=1)
X_test_gru = np.expand_dims(X_test, axis=1)

# Convert to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train_gru)
X_test_tensor = torch.FloatTensor(X_test_gru)
y_train_tensor = torch.FloatTensor(y_train.to_numpy()).reshape(-1, 1)
y_test_tensor = torch.FloatTensor(y_test.to_numpy()).reshape(-1, 1)

# === DATA LOADER ===
batch_size = 32
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# === GRU MODEL DEFINITION ===
class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size=64, dense_size=16):
        super(GRUModel, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, batch_first=True)
        self.fc1 = nn.Linear(hidden_size, dense_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(dense_size, 1)

    def forward(self, x):
        out, _ = self.gru(x)
        out = out[:, -1, :]  # Take last time step
        out = self.relu(self.fc1(out))
        return self.fc2(out)
```

```python
# === INSTANTIATE MODEL ===
input_size = X_train.shape[1]  # same as LSTM
model = GRUModel(input_size=input_size)

# === OPTIMIZER AND LOSS ===
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
 #=== TRAINING LOOP ===
epochs = 30
model.train()
for epoch in range(epochs):
    epoch_loss = 0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f"Epoch {epoch+1}/{epochs}, Loss: {epoch_loss:.4f}")
```

```python
# === EVALUATION ===
model.eval()
with torch.no_grad():
    y_pred_tensor = model(X_test_tensor)
    y_pred = y_pred_tensor.numpy()

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nGRU Model - MSE: {mse:.4f}")
print(f"GRU Model - R2 Score: {r2:.4f}")
```

### CNN-LSTM Hybrid Model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# === RESHAPE CORRECTLY ===
X_train_cnnlstm = np.expand_dims(X_train.to_numpy(), axis=1)  # (batch, 1, features)
X_test_cnnlstm = np.expand_dims(X_test.to_numpy(), axis=1)

# Convert to tensors
X_train_tensor = torch.FloatTensor(X_train_cnnlstm)
X_test_tensor = torch.FloatTensor(X_test_cnnlstm)
y_train_tensor = torch.FloatTensor(y_train.to_numpy()).reshape(-1, 1)
y_test_tensor = torch.FloatTensor(y_test.to_numpy()).reshape(-1, 1)
```

```python
# === CNN-LSTM MODEL ===
class CNNLSTMModel(nn.Module):
    def __init__(self, input_size, conv_channels=64, lstm_hidden=32, dense_size=16):
        super(CNNLSTMModel, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=conv_channels, kernel_size=2)
        self.relu = nn.ReLU()
        self.lstm = nn.LSTM(conv_channels, lstm_hidden, batch_first=True)
        self.fc1 = nn.Linear(lstm_hidden, dense_size)
        self.fc2 = nn.Linear(dense_size, 1)

    def forward(self, x):
        # x: (batch, 1, features)
        x = self.relu(self.conv1(x))        # --> (batch, conv_channels, new_seq_len)
        x = x.permute(0, 2, 1)              # --> (batch, new_seq_len, conv_channels)
        _, (hn, _) = self.lstm(x)
        x = hn[-1]
        x = self.relu(self.fc1(x))
        return self.fc2(x)
```

```python
# === INSTANTIATE MODEL ===
input_size = X_train.shape[1]
model = CNNLSTMModel(input_size=input_size)

# === LOSS AND OPTIMIZER ===
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# === TRAINING LOOP ===
epochs = 30
model.train()
for epoch in range(epochs):
    epoch_loss = 0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f"Epoch {epoch+1}/{epochs}, Loss: {epoch_loss:.4f}")
```

```python
# === EVALUATION ===
model.eval()
with torch.no_grad():
    y_pred_tensor = model(X_test_tensor)
    y_pred = y_pred_tensor.numpy()

# === METRICS ===
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nCNN-LSTM Model - MSE: {mse:.4f}")
print(f"CNN-LSTM Model - R2 Score: {r2:.4f}")
```

## Model Optimization

#### Hyperparameter Tuning

```python
import random
from itertools import product

# Define a random search space
param_grid = {
    'learning_rate': [0.001, 0.0005, 0.0001],
    'batch_size': [16, 32, 64],
    'hidden_size1': [32, 64, 128],
    'hidden_size2': [16, 32, 64],
    'dense_size': [8, 16, 32],
    'dropout_rate': [0.1, 0.2, 0.3]
}

# Sample N random combinations
def random_search(param_grid, n_iter=10):
    keys, values = zip(*param_grid.items())
    all_combinations = list(product(*values))
    random.shuffle(all_combinations)
    return [dict(zip(keys, comb)) for comb in all_combinations[:n_iter]]

param_samples = random_search(param_grid, n_iter=10)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, dense_size, dropout_rate):
        super(LSTMModel, self).__init__()
        self.lstm1 = nn.LSTM(input_size, hidden_size1, batch_first=True)
        self.dropout1 = nn.Dropout(dropout_rate)
        self.lstm2 = nn.LSTM(hidden_size1, hidden_size2, batch_first=True)
        self.dropout2 = nn.Dropout(dropout_rate)
        self.dense1 = nn.Linear(hidden_size2, dense_size)
        self.relu = nn.ReLU()
        self.dense2 = nn.Linear(dense_size, 1)

    def forward(self, x):
        x, _ = self.lstm1(x)
        x = self.dropout1(x)
        x, _ = self.lstm2(x)
        x = self.dropout2(x)
        x = x[:, -1, :]
        x = self.dense1(x)
        x = self.relu(x)
        x = self.dense2(x)
        return x
```

#### Early Stopping

```python
def train_model(params):
    model = LSTMModel(
        input_size=X_train_lstm.shape[2],
        hidden_size1=params['hidden_size1'],
        hidden_size2=params['hidden_size2'],
        dense_size=params['dense_size'],
        dropout_rate=params['dropout_rate']
    )

    optimizer = optim.Adam(model.parameters(), lr=params['learning_rate'])
    criterion = nn.MSELoss()

    # DataLoader
    X_train_tensor = torch.FloatTensor(X_train_lstm)
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    train_loader = DataLoader(train_dataset, batch_size=params['batch_size'], shuffle=True)

    # Early Stopping Setup
    patience = 10
    best_val_loss = float('inf')
    epochs_no_improve = 0
    best_model_state = None
    num_epochs = 100

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        train_loss /= len(train_loader)

        # Validation
        model.eval()
        with torch.no_grad():
            val_size = int(0.2 * len(X_train_tensor))
            X_val = X_train_tensor[:val_size]
            y_val = y_train_tensor[:val_size]
            val_outputs = model(X_val)
            val_loss = criterion(val_outputs, y_val).item()

        # Early Stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict()
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
        if epochs_no_improve >= patience:
            break

    model.load_state_dict(best_model_state)
    return model, best_val_loss
```

#### Model Evaluation Post-Optimization

```python
results = []

for i, params in enumerate(param_samples):
    print(f"Running config {i+1}/{len(param_samples)}: {params}")
    model, val_loss = train_model(params)
    
    # Evaluate on test set
    model.eval()
    with torch.no_grad():
        X_test_tensor = torch.FloatTensor(X_test_lstm)
        y_pred_tensor = model(X_test_tensor)
        y_pred = y_pred_tensor.numpy().flatten()
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

    results.append({
        'params': params,
        'val_loss': val_loss,
        'mse': mse,
        'r2': r2
    })
    print(f"Val Loss: {val_loss:.4f}, Test MSE: {mse:.4f}, R2: {r2:.4f}")
```

```python
# Find best model by validation loss or R2
best_result = sorted(results, key=lambda x: x['r2'], reverse=True)[0]

print("\nBest Model Parameters:")
print(best_result['params'])
print(f"Validation Loss: {best_result['val_loss']}")
print(f"Test MSE: {best_result['mse']}")
print(f"Test R2: {best_result['r2']}")


# Compare all R2 scores
r2_scores = [r['r2'] for r in results]
plt.plot(r2_scores, marker='o')
plt.title('R2 Scores for Different Hyperparameter Sets')
plt.xlabel('Configuration Index')
plt.ylabel('R2 Score')
plt.grid(True)
plt.show()
```
